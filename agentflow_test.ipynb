```python
# --- INJECTION SCRIPT (Corrected Path) ---
import os

# Path relative to where you are now
# You are likely in 'adk-samples/python/agents/travel-concierge'
# The agent file is inside the inner 'travel_concierge' folder
target_file = "travel_concierge/agent.py"

if not os.path.exists(target_file):
    print(f"‚ùå File not found: {target_file}")
    print(f"Current Directory: {os.getcwd()}")
    print(f"Files here: {os.listdir()}")
else:
    # 1. Read the original file
    with open(target_file, "r") as f:
        content = f.read()

    # 2. Check if already instrumented
    if "agentflow.sdk" not in content:
        print(f"üíâ Injecting AgentFlow into {target_file}...")

        # Add import at the top
        new_content = "from agentflow.sdk import instrument_agent\n" + content

        # Add instrumentation at the bottom
        new_content += "\n\n# --- AgentFlow Injection ---\n"
        new_content += "root_agent = instrument_agent(root_agent)\n"
        new_content += "print('‚úÖ AgentFlow Instrumentation Loaded')\n"

        # 3. Save the modified file
        with open(target_file, "w") as f:
            f.write(new_content)
        print("‚úÖ Injection Complete.")
    else:
        print("‚ö†Ô∏è File already instrumented.")
```
